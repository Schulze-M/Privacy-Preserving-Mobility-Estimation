# Plots the results of the evaluation of the Trie, alongside their standard deviation
The evaluation of the Trie is done in the `main.py` file. The results are saved in a CSV file in the `results` directory, which is then used to plot the results. \
To let `main.py` know that we want to plot the results, we need to set the `plot` argument parser variable to `True`. \
The `results` directory is created in the root directory of the project, if not present. \
`main.py --plot True`

## Why this IPYNB file?
This IPYNB file is used to plot the results of the evaluation of the Trie. \
The `main.py` script is not showing the plots, it only saves them. Use this IPYNB file to show the plots.
